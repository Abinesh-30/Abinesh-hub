K-means clustering

k = to the number of people you choose
its an unsupervised method that groups data into k cluster 

centroid

A centroid is the center point of a cluster in k-means clustering.

It’s basically the average position of all the data points in that cluster.

If your data has multiple features (e.g., height and weight), the centroid is the average height and average weight for that cluster.
