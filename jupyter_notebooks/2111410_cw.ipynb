Exercise 5: Find minimum and maximum values



```python
""" This code, defines a function called find_min_max which
    takes a list of numbers as input and then lets
    the user know the smallest and largest number  """

def find_min_max(numbers):
    smallest = min(numbers)
    biggest = max(numbers)
    return smallest, biggest

#example numbers
my_list = [4,9,2,15,7,1,100]

# Unpack the returned minium and maximum values
low_value, high_value = find_min_max(my_list)

# print the results
print("The smallest  value is:", low_value)
print("the largest Value is:", high_value)
```

Exercise 8: Is to solve the series
This code defines a functions, runs it to calculate the sum of the series.



```python
""" The code defines and runs a function to calculate the sum
of the following series:

x*5^2/(1+2!) + x*25^2/(2+3!)+ ... up to N terms 

"""
import math 

def solve_series(x, T):
   
    # Store the total of the series
    total = 0 

    # Loop  
    for i in range(1, T + 1):
        # Calcullate the numerator 
        numerater = x * (5 ** T) ** 2

        #calculate the denominator
        denominator = T + math.factorial(T + 1)

        #Add the current term to the total 
        total = total + (numerater / denominator)

    return total 

# Ask the user to enter x and N
x = int(input("Enter the value of x: "))
T = int(input("enter the number of terms: "))

# call the function and store the results
result = solve_series(x, T)

print("the sum of the series:", result)
```

Exercise 9: Load an image and create mirror reflections 



```python
""" This program reads a selected image and generates horizontal and vertical
    mirror reflection copies using skimage.io and matplotlib.pyplot libraries """
from skimage import io
import matplotlib.pyplot as plt 


# upload an image
image = io.imread("image.jpg")

#horizontal mirror reflection, flipping the image left to right side
h_flip = image[:, ::-1]

v_flip = image[::-1, :]

# displays the original imgage
plt.figure()
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()

# horizontal reflection 
plt.imshow(h_flip)
plt.title("horizontal reflection")
plt.axis("off")
plt.show()

#vertical reflection
plt.figure()
plt.imshow(v_flip)
plt.title("vertical refletion")
plt.axis("off")
plt.show()


```

Exercise 12 - Boxplots for the iris dataset


```python
""" using seaborn is loads the iris dataset and creates a seperate boxplots for each 
    numeriacal feature in the data set:
    sepal_length,sepal_width,petal_length,petal_width

 """
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

#display the first few rows of the dataset
print(iris.head())

numerical_columns = ["sepal_length","sepal_width","petal_length","petal_width"]

# create a boxplot 
for column in numerical_columns:
    plt.figure()
    plt.boxplot(iris[column])
    plt.title("Boxplot of" + column)
    plt.ylabel(column)
    plt.show()
```

Exercise 13 - Visualise restaurant  tips by gender and day


```python
""" This code  loads 'tips' data set, grouped the data by day,sex and time then 
    calculate the total tips for each group, then creates
    a chart using plotly express to visualise """
import seaborn as sns
import pandas as pd
import plotly.express as px

# load tips dataset from seaborn
tips = sns.load_dataset("tips")

tips_grouped = tips.groupby([ "day", "sex", "time"])["tip"].sum().reset_index()

#create the chart
chart = px.sunburst(
    tips_grouped,
    path = ["sex","day","time"],
    values= "tip",
    title ="total tips by day, sex and time"
)

#displays chart
chart.show()
```

Exercise 14 - Is using the previous exercise data to predict the bill amount.

```python
"""
This program uses linear regression model to predict the tips amount based on the whole bill
the data set is from the seaborn library.

"""
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression


#the data set
data = sns.load_dataset("tips")

#using the total bill to predict the tip
bill_data = data[["total_bill"]]
tip_data = data["tip"]

#create the model
reg = LinearRegression()
reg.fit(bill_data, tip_data)

example_bill = 28.50
example_input = pd.DataFrame({"total_bill": [example_bill]})

tip_prediction = reg.predict(example_input)

print("Bill amount:", example_bill)
print("predicted tip:", round(tip_prediction[0], 2))


```
