Exercise 8: Is to solve the series
This code defines a functions, runs it to calculate the sum of the series.



```python
# Import the math module to use factorial
import math 

def solve_series(x, T):
   
    # Store the total of the series
    total = 0 

    # Loop through each term from 1 to N 
    for n in range(1, T + 1):
        # Calcullate the numerator 
        # Pattern: 5, 25, 125, ...so this is 5^n
        numerater = x * (5 ** n) ** 2


        #calculate the denominator
        denominator = n + math.factorial(n + 1)

        #Add the current term to the total 
        total = total + (numerater / denominator)

    return total 

# Ask the user to enter x and N
x = int(input("Enter the value of x: "))
T = int(input("enter the number of terms: "))

# call the function and store the results
result = solve_series(x, T)

#print the final answer
print("the sume of the series is:", result)
```

Exercise 9: Load an image and create mirror reflections using NumPy

This code will load an image into a Numpy Array and will create a horizontal mirror reflection and vertical reflection and display it

```python
# imports libraries
from skimage import io
import matplotlib.pyplot as plt 


# upload an image
image = io.imread("image.jpg")

#horizontal mirror reflection, flipping the image left to right side
h_flip = image[:, ::-1]

v_flip = image[::-1, :]

# displays the original imgage
plt.figure()
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()

# dsplay the horizontal reflection 
plt.imshow(h_flip)
plt.title("horizontal reflection")
plt.axis("off")
plt.show()

# display the vertical reflection
plt.figure()
plt.imshow(v_flip)
plt.title("vertical refletion")
plt.axis("off")
plt.show()


```

Exercise 12 - Boxplots for the iris dataset

in this code it will download the iris dataset from the seaborn library and uses matplotlib to generate boxplots for each numerical column.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load the iris dataset from seaborn
iris = sns.load_dataset("iris")

#display the first few rows of the dataset
print(iris.head())

#select the numerical columns from the dataset
numerical_columns = ["sepal_length","sepal_width","petal_length","petal_width"]

# create a boxplot for each numerical column
for column in numerical_columns:
    plt.figure()
    plt.boxplot(iris[column])
    plt.title("Boxplot of" + column)
    plt.ylabel(column)
    plt.show()
```

Exercise 13 - Visualise restraunt tips by gender and day

This program uses the tips dataset from the seabron library and creats a sunburst chart using plotly.express


```python
#import libraries
import seaborn as sns
import pandas as pd
import plotly.express as px

# load the tips dataset from seaborn
tips = sns.load_dataset("tips")

#display the first few rows of the dataset
print(tips.head())

#group the dta by sex, day and time then calculate the total tips
tips_grouped = tips.groupby(["sex","day","time"])["tip"].sum().reset_index()

#print the grouped data
print(tips_grouped)

#create the sunburst chart
fig = px.sunburst(
    tips_grouped,
    path=["sex","day","time"],
    values="tip",
    title="Total tips grouped by sex, day and time"
)

#show the chart
fig.show()
```
