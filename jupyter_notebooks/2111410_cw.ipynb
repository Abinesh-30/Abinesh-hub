Exercise 5: Find minimum and maximum values
For this exercise it defines a function that accepts an iterable of numbers and returns the minimum and maximum values


```python
def find_min_max(numbers):
    smallest = min(numbers)
    biggest = max(numbers)
    return smallest, biggest

#example of iterable numbers
my_list = [4,9,2,15,7,1,100]

# Unpack the returned minium and maximum values
low_value, high_value = find_min_max(my_list)

# print the results
print("The smallest  value is:", low_value)
print("the largest Value is:", high_value)
```

Exercise 6: Guess the number game 
For this exercise, it chooses a random integer between 1 and 10 and user must keep guessing until the correct number is entered.

```python

import random

#  random secret number between 1 and 10
secret_number = random.randint(1,10)
#set an initial guess value
guess = 0

# keep asking asking to guess until the number is correct
while guess != secret_number:
    guess = int(input(" guess between 1 and 10"))


    if guess > secret_number:
        print("Too high! enter again.")
    elif guess < secret_number:
        print("Too low! Try agian.")
    else:
        print("Congratulation!!!!!!!")
        
```

Exercise 9: Load an image and create mirror reflections using NumPy

This code will load an image into a Numpy Array and will create a horizontal mirror reflection and vertical reflection and display it

```python
# imports libraries
from skimage import io
import matplotlib.pyplot as plt 


# upload an image
image = io.imread("image.jpg")

#horizontal mirror reflection, flipping the image left to right side
h_flip = image[:, ::-1]

v_flip = image[::-1, :]

# displays the original imgage
plt.figure()
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()

# horizontal reflection 
plt.imshow(h_flip)
plt.title("horizontal reflection")
plt.axis("off")
plt.show()

#vertical reflection
plt.figure()
plt.imshow(v_flip)
plt.title("vertical refletion")
plt.axis("off")
plt.show()


```

Exercise 12 - Boxplots for the iris dataset

in this code it will download the iris dataset from the seaborn library and uses matplotlib to generate boxplots for each numerical column.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load the iris dataset from seaborn
iris = sns.load_dataset("iris")

#display the first few rows of the dataset
print(iris.head())

#select the numerical columns from the dataset
numerical_columns = ["sepal_length","sepal_width","petal_length","petal_width"]

# create a boxplot for each numerical column
for column in numerical_columns:
    plt.figure()
    plt.boxplot(iris[column])
    plt.title("Boxplot of" + column)
    plt.ylabel(column)
    plt.show()
```

Exercise 13 - Visualise restraunt tips by gender and day

This program uses the tips dataset from the seabron library and creats a sunburst chart using plotly.express


```python

import seaborn as sns
import pandas as pd
import plotly.express as px

# load tips dataset from seaborn
tips = sns.load_dataset("tips")

#group the total dta by sex, day and time then calculate the total tips
tips_grouped = tips.groupby([ "day", "sex", "time"])["tip"].sum().reset_index()

#create the chart
chart = px.sunburst(
    tips_grouped,
    path = ["sex","day","time"],
    values= "tip",
    title ="total tips by day, sex and time"
)

#show the chart
chart.show()
```

Exercise 14 - Is using the previous exercise data to predict the bill amount.

```python
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression


#the data set
data = sns.load_dataset("tips")

#using the total bill to predict the tip
bill_data = data[["total_bill"]]
tip_data = data["tip"]

#create the model
reg = LinearRegression()
reg.fit(bill_data, tip_data)

example_bill = 28.50
example_input = pd.DataFrame({"total_bill": [example_bill]})

tip_prediction = reg.predict(example_input)

print("Bill amount:", example_bill)
print("predicted tip:", round(tip_prediction[0], 2))


```
