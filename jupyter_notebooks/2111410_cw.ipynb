Exercise 1: Prime number checker

This Exercise asks the user to enter a number and checks whether it's a prime number.
 


```python
# Ask the user to enter a number
num = int(input("Enter any number"))

# If the number is less than or equal to 1, it's not prime
if num<= 1:
    print("This number is not prime")
else:
    # Assume the number is prime unless proven otherwise
    is_prime = True

    #check if the number iss divisible by any number from 2 to -1
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    # print the results
    if is_prime:
        print("This number is a prime number")
    else:
        print("This number isn't prime")
```

Exercise 2: Fibonacci Sequence
For this exercise ask the user how many numbers in fibonacci set to be calculated and display it.

```python
# Fibonacci sequence

#Asks the user how many numbers to display
n = int(input("Enter how many numbers in fibonacci set to be calculated and displayed?"))

#Starting values of the Fibonacci sequence
a = 0 
b = 1

# Repeat n times
for i in range(n):
    print(a)

    # calculate the next number in the fibonacci sequence
    c = a + b
    a = b
    b = c
```

Exercise 3: printing pattern
For this exercise print the star pattern

The first half of the pattern starts from 5 stars, then decreases to 1 star
The second half of the pattern does the opposite and increases back to 5 stars

```python
# printing pattern

# the first section from 5 star to 1 star
for i in range(5, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print() # moves to the next line

# Second half of the pattern
for i in range(2, 6):
    for j in range(i):
        print("*", end=" ")
    print()
```

Exercise 4: Reverse a string  
For this Exercise it defines a function which asks the user to enter a string and then prints it in reverse order.


```python
# Function to reverse a string 

def reverse_string():

    #Ask the user to enter a string
    text =  input("Enter a string: ")

    # Reverse the string using slicing
    reversed_text = text[::-1]

    # print the reversed string
    print("Reveresed string:", reversed_text)
    
# call the function
reverse_string()
```

Exercise 5: Find minimum and maximum values
For this exercise it defines a function that accepts an iterable of numbers and returns the minimum and maximum values


```python
def find_min_max(numbers):
    return min(numbers), max(numbers)

#example of iterable numbers
values = [4,9,2,15,7,1,100]

# Unpack the returned minium and maximum values
min_value, max_value = find_min_max(values)

# print the results
print("The min value is:", min_value)
print("The max value is:", max_value)
```

Exercise 6: Guess the number game 
For this exercise, it chooses a random integer between 1 and 10 and user must keep guessing until the correct number is entered.

```python
#Guess the number Game

import random

#generate a random secret number between 1 and 10
secret_number = random.randint(1,10)

#set an initial guess value
guess = 0

# keep asking asking to guess until the number is correct
while guess != secret_number:
    guess = int(input("Enter your guess between 1 and 10"))

# compare the guessed number with the secret number
    if guess > secret_number:
        print("Too high! Try again.")
    elif guess < secret_number:
        print("Too low! Try agian.")
    else:
        print("Congratulations! youve guessed correctly")
        
```

Exercise 7: Password checker with a control loop
In this exercise the code keeps asking the user to enter a password until the correct password is entered

The code uses an infinite while true loop and exits using the break keyword

```python
#store the correct password
correct_password = "password"

#infinite loop
while True:
    #ask the user to enter a password
    user_password = input("Enter the password:")


    # check if the password is correct
    if user_password == correct_password:
        print("Access Granted!")
        break
    else:
        print(" Incorrect password. Please try again.")
```
