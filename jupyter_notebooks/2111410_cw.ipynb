```python
# Ask the user to enter a number
num = int(input("Enter any number"))

# check if the number is prime
if num<= 1:
    print("This number is not prime")
else:
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print("This number is a prime number")
    else:
        print("This number isn't prime")
```

```python
# Fibonacci sequence

n = int(input("Enter how many numbers in fibonacci set to be calculated and displayed?"))

a = 0 
b = 1

for i in range(n):
    print(a)
    c = a + b
    a = b
    b = c
```

```python
# printing pattern

# the first section from 5 star to 1 star
for i in range(5, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()

# Second half of the pattern
for i in range(2, 6):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
# Function to reverse a string 

def reverse_string():
    text =  input("Enter a string: ")
    reversed_text = text[::-1]
    print("Reveresed string:", reversed_text)

reverse_string()
```
